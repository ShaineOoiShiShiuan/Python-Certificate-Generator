**Add a certificate id**

How are we going to generate a unique certificate ID ?

We are going to use a UNIX time stamp

```python
pd.Timestamp.now()
pd.Timestamp.now().timestamp() #UNIX timestamp
int(pd.Timestamp.now().timestamp())
```

```python
def certificate_generator(name,courseName,courseLevel,date):

  pdf_file_name = "/content/drive/MyDrive/Python Certificate /Certificate/ " + name + "-" + courseName + "-" + courseLevel + ".pdf"
#create a black canvas or a blank paper virtually using the function canvas.Canvas()
  c = canvas.Canvas(pdf_file_name, pagesize= landscape(A4))

#Step 2 : Paste my template on my blank sheet of paper
#I am going to use c.drawimage
#1st input ---> I am retriving my template
#2nd input ---> x-axis (Cordinate start from 0)
#3rd input ---> y-axis (Cordinate start from 0)
#4th input ---> width ---> A4[1]
#5th input ---> height ---> A4[0]

  c.drawImage("/content/drive/MyDrive/Python Certificate /Certificate template.jpg", 0 ,0, width = A4[1], height= A4[0])

  center_x = c._pagesize[0] / 2
  center_y = c._pagesize[1] / 2

  c.setFont('Lora-Bold',30)

  c.drawCentredString(center_x, center_y-50, name)
  #Populate the template
  c.setFont('Lora-Bold',30)
  c.drawCentredString(center_x,center_y-110,courseName +"-" + courseLevel)

  c.setFont('Lora-Bold',17)
  c.drawCentredString(center_x+190,center_y-160,date)
  #Step 4:

  cert_id="Cert_ID:" + str(int(pd.Timestamp.now().timestamp()))
  c.setFont("Lora-Regular",12)
  c.drawCentredString(center_x+266, center_y-230,cert_id.upper())

  c.save()
```
